```python
# Operations in dataframes

import pandas as pd
import numpy as np

# Create DataFrame
data = {
    "ID": [101, 102, 103, 104, 105],
    "Name": ["Amit", "Bobby", "Charlie", "Danush", "Emily"],
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Age": [25, 30, 28, 35, 29],
    "Salary": [40000, 60000, 55000, 70000, 45000]
}

df = pd.DataFrame(data)

# Display DataFrame
print("\nOriginal DataFrame:")
print(df)

# Basic Information
print("\nHead:")
print(df.head())

print("\nTail:")
print(df.tail())

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nInfo:")
print(df.info())

print("\nDescribe:")
print(df.describe())

# Selecting columns
print("\nName column:")
print(df["Name"])

print("\nMultiple columns:")
print(df[["Name", "Salary"]])

# Filtering
print("\nSalary > 50000:")
print(df[df["Salary"] > 50000])

# loc and iloc
print("\nloc:")
print(df.loc[1:3])

print("\niloc:")
print(df.iloc[0:3, 1:4])

# Adding columns
df["Bonus"] = df["Salary"] * 0.1
print("\nAfter Adding Bonus:")
print(df)

# Updating values
df.loc[df["Name"] == "Bob", "Salary"] = 65000
print("\nUpdated Salary:")
print(df)

# Sorting
print("\nSorted by Salary:")
print(df.sort_values(by="Salary", ascending=False))

# Unique values
print("\nUnique Departments:")
print(df["Department"].unique())

print("\nNumber of Unique Departments:")
print(df["Department"].nunique())

# Value counts
print("\nDepartment Counts:")
print(df["Department"].value_counts())

# Groupby
print("\nAverage Salary Department-wise:")
print(df.groupby("Department")["Salary"].mean())

# Aggregation
print("\nAggregations:")
print(df["Salary"].agg(["min", "max", "mean", "sum"]))

# Missing values
print("\nMissing Values:")
print(df.isnull())

print("\nMissing Values Count:")
print(df.isnull().sum())

# Fill missing values
df.fillna(0, inplace=True)

# Duplicate rows
print("\nDuplicate Rows:")
print(df.duplicated())

df.drop_duplicates(inplace=True)

# Rename columns
df.rename(columns={"Age": "Employee_Age"}, inplace=True)

# Drop column
df.drop(columns=["Bonus"], inplace=True)

# Correlation
print("\nCorrelation:")
print(df.select_dtypes(include=np.number).corr())

# Pivot Table
print("\nPivot Table:")
print(pd.pivot_table(df,
                     values="Salary",
                     index="Department",
                     aggfunc="mean"))

# Statistical functions
print("\nMaximum Salary:", df["Salary"].max())
print("Minimum Salary:", df["Salary"].min())
print("Mean Salary:", df["Salary"].mean())
print("Median Salary:", df["Salary"].median())
print("Standard Deviation:", df["Salary"].std())

# Sampling
print("\nRandom Sample:")
print(df.sample(2))

# Index operations
print("\nIndex:")
print(df.index)

# Reset index
df.reset_index(drop=True, inplace=True)

# Export to CSV
df.to_csv("employee.csv", index=False)

# Read CSV
new_df = pd.read_csv("employee.csv")

print("\nFinal DataFrame:")
print(new_df)
```
