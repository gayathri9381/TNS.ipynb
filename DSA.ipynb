```python
# Reverse a string using stack
s = "HELLO"
stack = []

for ch in s:
    stack.append(ch)

reverse = ""

while stack:
    reverse += stack.pop()

print(reverse)
```

```python
# sort a stack
stack = [5,2,8,1]

stack.sort()

print(stack)
```

```python
# check palindrom using stack
s="MADAM"

stack=list(s)

rev=""

while stack:
    rev += stack.pop()

if s==rev:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
# reverse a linkedlist
prev = None
current = head

while current:
    next_node = current.next
    current.next = prev
    prev = current
    current = next_node

head = prev
```

```python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Reverse function
def reverse_linked_list(head):
    prev = None
    current = head

    while current:
        next_node = current.next      # Store next node
        current.next = prev           # Reverse link
        prev = current                # Move prev
        current = next_node           # Move current

    return prev


# Function to print linked list
def print_list(head):
    temp = head
    while temp:
        print(temp.data, end=" -> ")
        temp = temp.next
    print("NULL")


# Create linked list
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

print("Original Linked List:")
print_list(head)

# Reverse the linked list
head = reverse_linked_list(head)

print("Reversed Linked List:")
print_list(head)
```

```python
# create a tree
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Create tree
root = Node(10)

root.left = Node(20)
root.right = Node(30)

root.left.left = Node(40)
root.left.right = Node(50)

root.right.left = Node(60)
root.right.right = Node(70)
```

```python
# Inorder traversal
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Create tree
root = Node(10)

root.left = Node(20)
root.right = Node(30)

root.left.left = Node(40)
root.left.right = Node(50)

root.right.left = Node(60)
root.right.right = Node(70)

# Inorder Traversal
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)

print("Inorder Traversal:")
inorder(root)
```

```python
# postorder traversal
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data, end=" ")

postorder(root)
```

```python
# Preorder traversal
def preorder(root):
    if root:
        print(root.data,end=" ")
        preorder(root.left)
        preorder(root.right)
```

```python
# height of a tree
def height(root):
    if root is None:
        return 0

    return 1 + max(height(root.left), height(root.right))

print(height(root))
```
