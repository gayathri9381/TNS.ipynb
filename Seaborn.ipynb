## Seaborn?

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive, informative, and statistical graphics with less code.

##Difference Between Matplotlib and Seaborn

| Feature | Matplotlib | Seaborn |
|----------|--------|----------|
| Level | Low-level visualization library | High-level visualization library |
| Built On | Base plotting library | Built on Matplotlib |
| Syntax | More code required | Less code required |
| Statistical Plots | Limited support | Excellent support |
| DataFrame Support | Requires manual handling | Directly works with Pandas DataFrames |
| Customization | Highly customizable | Easy customization |
| Learning Curve | Slightly difficult | Beginner-friendly |



```python
#Matplotlib Example
import matplotlib.pyplot as plt

students = ['A','B','C','D']
marks = [78,85,92,88]

plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#Seaborn Example
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'Students':['A','B','C','D'],
    'Marks':[78,85,92,88]
})

sns.barplot(x='Students', y='Marks', data=df)
plt.title("Student Marks")
plt.show()
```

## 3. List of Commonly Used Functions in Seaborn

1. `sns.scatterplot()` – Creates scatter plots.  
2. `sns.lineplot()` – Creates line plots.  
3. `sns.barplot()` – Shows average values with confidence intervals.  
4. `sns.countplot()` – Counts frequency of categorical data.  
5. `sns.boxplot()` – Creates box-and-whisker plots.  
6. `sns.violinplot()` – Combines boxplot and KDE for distribution.  
7. `sns.heatmap()` – Creates heatmaps for correlation matrices or tabular data.  
8. `sns.pairplot()` – Pairwise relationships in dataset.  
9. `sns.distplot()` / `sns.histplot()` – Distribution of variables.  
10. `sns.regplot()` – Regression line with scatter plot.  
11. `sns.catplot()` – Plotting categorical data.

##Implementation

```python
# Seaborn Implementation - All in One Code

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
tips = sns.load_dataset("tips")

# 1. Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="time")
plt.title("Scatter Plot - Total Bill vs Tip")
plt.show()

# 2. Line Plot
plt.figure(figsize=(6,4))
sns.lineplot(x="size", y="total_bill", data=tips, ci=None)
plt.title("Line Plot - Total Bill vs Size")
plt.show()

# 3. Bar Plot
plt.figure(figsize=(6,4))
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Bar Plot - Average Total Bill per Day")
plt.show()

# 4. Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x="day", data=tips, hue="sex")
plt.title("Count Plot - Customers per Day")
plt.show()

# 5. Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Box Plot - Total Bill Distribution")
plt.show()

# 6. Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
plt.title("Violin Plot - Distribution by Day and Gender")
plt.show()

# 7. Heatmap (Correlation Matrix)
plt.figure(figsize=(6,4))
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# 8. Pairplot
sns.pairplot(tips, hue="sex")
plt.suptitle("Pairplot of Tips Dataset",y=1.02)
plt.show()

# 9. Distribution Plot (Histogram + KDE)
plt.figure(figsize=(6,4))
sns.histplot(tips["total_bill"], kde=True, bins=20)
plt.title("Distribution of Total Bill")
plt.show()

# 10. Regression Plot
plt.figure(figsize=(6,4))
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot - Tip vs Total Bill")
plt.show()
```

Perform the following:
- Display dataset information.
- Create a scatter plot between petal length and petal width.
- Color points based on species.
- Create a pair plot.
- Identify which species appears most distinct.

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset
iris = sns.load_dataset("iris")

# 1. Display dataset information
print("First 5 rows of the dataset:")
print(iris.head())

print("\nDataset Information:")
print(iris.info())

print("\nStatistical Summary:")
print(iris.describe())

# 2. Create a scatter plot between petal length and petal width
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species",
    palette="Set1",
    s=80
)

plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title="Species")
plt.show()

# 3. Create a pair plot
sns.pairplot(iris, hue="species", palette="Set1")
plt.show()

# 4. Identify the most distinct species
print("\nObservation:")
print("Setosa appears to be the most distinct species because")
print("its measurements are clearly separated from Versicolor and Virginica.")
```
